```python
from selenium import webdriver 
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from io import BytesIO
from PIL import Image
from time import sleep

def get_obs(driver, save=True):
    screenshot = driver.get_screenshot_as_png()
    image = Image.open(BytesIO(screenshot))
    image.save('current_view.png') if save else None
    return image

def step(value):
    sleep(1)
    actions = ActionChains(driver) 
    actions.send_keys(value)
    actions.perform()
    obs = get_obs(driver)
    return obs

options = webdriver.ChromeOptions()
options.add_extension('./vimium.crx')
# options.add_argument('headless')
driver = webdriver.Chrome(options=options) 
driver.set_window_size(768, 768)
start_link = "http://ec2-3-131-244-37.us-east-2.compute.amazonaws.com:3000/"
driver.get(start_link)
actions = ActionChains(driver) 
```
